```python
# import library
from tensorflow import keras
import pandas as pd
import numpy  as np
import joblib
from tensorflow.keras.models import load_model
```

```python
# load preprocessor
with open('preprocessor.pkl','rb') as file_1:
    preprocessor = joblib.load(file_1)
```

```python
# load ANN model
model = load_model('model_func_2_model.h5')
```

# Data Loading

```python
data_infer = pd.read_csv('h8dsft_P2M1_inference_Mathilda.csv')
```

```python
data_infer.head(10)
```

```python
data_new = data_infer.copy()
data_new = data_infer.drop(['Unnamed: 0'],axis=1)
data_new.head(10)
```

# Model Inference

```python
# preprocess data before feedforwarding into the model
preprocessed_data = preprocessor.transform(data_new)
```

```python
# predict result
predictions = model.predict(preprocessed_data)
```

```python
# convert prediction probability to each appropriate target class
pred = np.where(predictions>0.5,1,0)
```

```python
# show data infer + prediction
data_new['churn_risk_score'] = pred
data_new.head(10)
```
